# how to apply a function on a dataframe

<UI>
    <li>applymap
        <li>apply

```python
df_movie = pd.DataFrame({"movie_1":[5,4,3,5,1],"movie_2":[2,3,4,5,1]},index=["Tom","Jeff","Peter","Ram","Paul"])
```

```python
df_movie
```

```python
# we are defining a function to assign corresponding grades for the ratings
def movie_grade(rating):
    if rating == 5:
        return "A"
    if rating == 4:
        return "B"
    if rating == 3:
        return "C"
    else:
        return "F"
```

```python
movie_rating = 4

if movie_rating == 5:
    print("this movie is great")
else:
         print("The movie is average")
```

# Now lets see how we can apply this function on our dataframe

```python
df_movie.applymap(movie_grade) # apply functions on dataframe
```

```python
df_movie['movie_1'].apply(movie_grade) # apply function only on specific columns
```
